```python
# Case 1: K=5 (total 10 samples for each task)
K = 5
score_dict  = {'SCORE':multivariate_Uniform_score,'dim':2, 'null_param':0}
tasks = Oscillatory_Task_Distribution(2, 4, 6, 0.4, 0.6, 0, 1)
```

```python
# MCV
inlr = 0.01
instep=1
mCV5 = MetaNeuralCV('oscillatory', MetaNeuralCVModel, D_in=2, h_dims=[80] * 2, init_val=torch.zeros(1), weight_decay=5.e-6,\
                   tasks=tasks, inner_optim=torch.optim.Adam, inner_lr=inlr, meta_lr=2e-3, K=K, inner_steps=instep,\
                   tasks_per_meta_batch=5,\
                   **score_dict)
mCV5.main_loop(num_iterations=4000)
mCV5.test(1000, inner_optim=torch.optim.Adam, inner_steps=instep)
```

```python
#NCV
out5 = torch.zeros(1000,2)
for i in range(1000):
    if (i+1)%100==0:
        print('{}/{}'.format(i+1,1000))
    nCV = NeuralCV('oscillatory', NeuralCVModel, D_in=2, h_dims=[80] * 2, init_val=torch.zeros(1), weight_decay=5.e-6, \
                   tasks=tasks, optim=torch.optim.Adam, lr=2e-3, K=K,\
                   **score_dict)
    __ = nCV.train_val(num_epochs=20, batch_size=5, verbose=False)
    out5[i,0] = __[0]
    out5[i,1] = __[1]
```

```python
import pandas as pd
no_tasks = 1000
set_of_ss = [5, 10, 20, 50,100]
no_ss = len(set_of_ss)
```

```python

for i in range(no_ss):
    if set_of_ss[i] == 5:
        mCV_log = mCV5
        nCV_log = out5
    elif set_of_ss[i] == 10:
        mCV_log = mCV10
        nCV_log = out10
    elif set_of_ss[i] == 20:
        mCV_log = mCV20
        nCV_log = out20
    elif set_of_ss[i] == 50:
        mCV_log = mCV50
        nCV_log = out50
    elif set_of_ss[i] == 100:
        mCV_log = mCV100
        nCV_log = out100
    cur_MCV_data = list(zip(np.array(mCV_log.log_test[0]['Abserr_CVests']), np.repeat('MCV', no_tasks), np.repeat("m={}".format(set_of_ss[i]), no_tasks)))
    cur_MCV_data_df = pd.DataFrame(data=cur_MCV_data, columns=['est_abserr', 'method', 'sample_size'])
    if i == 0:
        MCV_data_df = cur_MCV_data_df
    if i >= 1:
        MCV_data_df = MCV_data_df.append(cur_MCV_data_df)


    # NCV
    cur_NCV_data = list(zip(nCV_log[:,1].numpy(), np.repeat('NCV', no_tasks), np.repeat("m={}".format(set_of_ss[i]), no_tasks)))
    cur_NCV_data_df = pd.DataFrame(data= cur_NCV_data, columns=['est_abserr', 'method', 'sample_size'])
    if i == 0:
        NCV_data_df = cur_NCV_data_df
    if i >= 1:
        NCV_data_df= NCV_data_df.append(cur_NCV_data_df)

    # MC
    cur_MC_data =  list(zip(nCV_log[:,0].numpy(), np.repeat('MC', no_tasks), np.repeat("m={}".format(set_of_ss[i]), no_tasks)))
    cur_MC_data_df = pd.DataFrame(data=cur_MC_data,columns=['est_abserr', 'method', 'sample_size'])
    if i == 0:
        MC_data_df = cur_MC_data_df
    if i >= 1:
        MC_data_df= MC_data_df.append(cur_MC_data_df)
```

```python
# Merge into one giant dataset
Oscillatory_DF_C = MCV_data_df.append([MC_data_df, NCV_data_df])
```
