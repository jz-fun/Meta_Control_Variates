```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget
from src.score_funcs import *
from Oscillatory_C.MCV_oscillatory_c import *
from Oscillatory_C.Oscillatory_Tasks import *
from Oscillatory_C.NCV_oscillatory_c import *
```

```python
# Experiments
nums_insteps = [1,3,5,7,10]

results_mcv5_invest_insteps_C = torch.zeros(len(nums_insteps),2)
results_mc5_invest_insteps_C = torch.zeros(len(nums_insteps),2)
for j in range(0,len(nums_insteps),1):

    # Fix dim and sample size
    dim=2
    K = 5

    score_dict  = {'SCORE':multivariate_Uniform_score,'dim':dim, 'null_param':0}
    tasks = Oscillatory_Task_Distribution(dim, 4, 6, 0.4, 0.6, 0, 1) #dim=2
    n_tests = 1000

    if j == 0:
        inlr = 1e-2
    else:
        inlr = 1e-2/(50*(1+j)) # alpha propto O(1/L)


    # Change number of inner steps
    insteps=nums_insteps[j]
    mCV = MetaNeuralCV('oscillatory',MetaNeuralCVModel, D_in=dim,  h_dims=[80] * 2, init_val=torch.zeros(1), weight_decay= 5.e-6, \
                 tasks=tasks, inner_optim = torch.optim.Adam, inner_lr=inlr, meta_lr=2e-3, K=K, inner_steps=insteps, tasks_per_meta_batch=5,\
                       **score_dict)
    mCV.main_loop(num_iterations=4000)

    if j == 1:
        fig = plt.figure()
    plt.plot(mCV.meta_losses)
    plt.show()

    # Meta testing
    mCV.test(n_tests, inner_optim = torch.optim.Adam,inner_steps=insteps)

    results_mcv5_invest_insteps_C[j-1,0] = torch.tensor(np.mean(np.array(mCV.log_test[0]['Abserr_CVests']))).detach().clone()
    results_mcv5_invest_insteps_C[j-1,1] = torch.tensor(np.std(np.array(mCV.log_test[0]['Abserr_CVests']))).detach().clone()
    results_mc5_invest_insteps_C[j-1,0] = torch.tensor(np.mean(np.array(mCV.log_test[0]['Abserr_MC_2m_ests']))).detach().clone()
    results_mc5_invest_insteps_C[j-1,1] = torch.tensor(np.std(np.array(mCV.log_test[0]['Abserr_MC_2m_ests']))).detach().clone()

    print('{}: mc:{}, mcv:{}'.format(j, results_mc5_invest_insteps_C[j-1],results_mcv5_invest_insteps_C[j-1] ))

```
