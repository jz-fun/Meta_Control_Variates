```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
%matplotlib widget
from src.score_funcs import *
from Oscillatory_C.MCV_oscillatory_c import *
from Oscillatory_C.Oscillatory_Tasks import *
from Oscillatory_C.NCV_oscillatory_c import *
```

```python
# Experiments
results_ncv5 = torch.zeros(7,2)
results_mcv5 = torch.zeros(7,2)
results_mc5 = torch.zeros(7,2)
time_metatrain = []
time_metatest = []
time_ncv = []
for j in range(1,8,1):
    K = 5
    dim=j
    score_dict  = {'SCORE':multivariate_Uniform_score,'dim':dim, 'null_param':0}
    tasks = Oscillatory_Task_Distribution(dim, 4, 6, 0.4, 0.6, 0, 1)
    n_tests = 1000

    inlr = 1e-2
    insteps=1

    mCV = MetaNeuralCV('oscillatory',MetaNeuralCVModel, D_in=dim,  h_dims=[80] * 2, init_val=torch.zeros(1), weight_decay= 5.e-6,\
                 tasks=tasks, inner_optim = torch.optim.Adam, inner_lr=inlr, meta_lr=2e-3, K=K, inner_steps=insteps, tasks_per_meta_batch=5,\
                       **score_dict)
    start_train_time = time.time()
    mCV.main_loop(num_iterations=4000)
    end_train_time = time.time()
    if j <= 2: # for j > 2, I use a large number of MC means to get the true vals so the time is not true for j >2.
        time_metatrain.append(end_train_time - start_train_time)
    print(time_metatrain)

    if j == 1:
        fig = plt.figure()
    plt.plot(mCV.meta_losses)
    plt.show()

    # Meta testing
    start_test_time = time.time()
    mCV.test(n_tests, inner_optim = torch.optim.Adam,inner_steps=insteps)
    end_test_time = time.time()
    if j <= 2: # for j > 2, I use a large number of MC means to get the true vals so the time is not true for j >2.
        time_metatest.append(end_test_time - start_test_time)
    print(time_metatest)

    # print( torch.tensor(np.mean(np.array(mCV.log_test[0]['Abserr_CVests']))).detach().clone())
    results_mcv5[j-1,0] = torch.tensor(np.mean(np.array(mCV.log_test[0]['Abserr_CVests']))).detach().clone()
    results_mcv5[j-1,1] = torch.tensor(np.std(np.array(mCV.log_test[0]['Abserr_CVests']))).detach().clone()
    results_mc5[j-1,0] = torch.tensor(np.mean(np.array(mCV.log_test[0]['Abserr_MC_2m_ests']))).detach().clone()
    results_mc5[j-1,1] = torch.tensor(np.std(np.array(mCV.log_test[0]['Abserr_MC_2m_ests']))).detach().clone()


    n_replica=n_tests
    out = torch.zeros(n_replica,2)
    start_ncv_time = time.time()
    for i in range(n_replica):
        nCV = NeuralCV('oscillatory', NeuralCVModel, D_in=j, h_dims=[80] * 2, init_val=torch.zeros(1), weight_decay=5.e-6,\
                        tasks=tasks, optim=torch.optim.Adam, lr=2e-3, K=K,\
                        **score_dict)
        __ = nCV.train_val(num_epochs=20, batch_size=5, verbose=False)
        out[i,0] = torch.tensor(__[0]).detach().clone()
        out[i,1] =torch.tensor(__[1]).detach().clone()
    end_ncv_time = time.time()
    if j <= 2: # for j > 2, I use a large number of MC means to get the true vals so the time is not true for j >2.
        time_ncv.append(end_ncv_time-start_ncv_time)
    print(time_ncv)

    results_ncv5[j-1,0] = torch.tensor(out.mean(0)[1]).detach().clone()
    results_ncv5[j-1,1] = torch.tensor(out.std(0)[1]).detach().clone()

    print('{}: mc:{}, ncv:{}, mcv:{}'.format(j, results_mc5[j-1],results_ncv5[j-1],results_mcv5[j-1] ))
```

```python
# Save results
torch.save(results_mcv5, 'results_mcv5_C.pt')
torch.save(results_mc5, 'results_mc5_C.pt')
torch.save(results_ncv5, 'results_ncv5_C.pt')
```
